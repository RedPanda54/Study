**Quick vs Merge vs Heap**

```python
import matplotlib.pyplot as plt

#  (n)
n_values = [100, 500, 1000, 5000, 10000, 50000, 100000, 200000, 500000]

# Quick Sort +,  
quick_sort_data = [1137, 8711, 20553, 123864, 238093, 1235334, 2418486, 5117242, 12264796]
quick_sort_time = [0.00017, 0.0013, 0.0031, 0.019, 0.035, 0.27, 0.44, 1.07, 3.34]

# Merge Sort +,  
merge_sort_data = [1211, 8365, 18684, 117060, 254107, 1502659, 3205367, 6810267, 18312299]
merge_sort_time = [0.00029, 0.0019, 0.0042, 0.026, 0.057, 0.42, 0.93, 2.03, 5.9]

# Heap Sort +,  
heap_sort_data = [1333, 9439, 21372, 136106, 297034, 1776502, 3801612, 8103531, 21902425]
heap_sort_time = [0.00027, 0.0019, 0.0046, 0.03, 0.088, 0.56, 1.25, 3.38, 8.29]

# Time Sort  
tim_sort_time = [9.81e-06, 4.61e-05, 9.52e-05, 0.0006, 0.0012, 0.007, 0.04, 0.03, 0.14]

#  
fig, ax1 = plt.subplots()

ax1.plot(n_values, quick_sort_data, marker='o', label='Quick Sort(C+S)', color='b')
ax1.plot(n_values, merge_sort_data, marker='o', label='Merge Sort(C+S)', color='g')
ax1.plot(n_values, heap_sort_data, marker='o', label='Heap Sort(C+S)', color='m')
ax1.set_xlabel('Input size (n)')
ax1.set_ylabel('comparison + swap', color='k')
ax1.set_yscale('log')  # +    

ax2 = ax1.twinx()
ax2.plot(n_values, quick_sort_time, marker='o', label='Quick Sort(Time)', color='r')
ax2.plot(n_values, merge_sort_time, marker='o', label='Merge Sort(Time)', color='y')
ax2.plot(n_values, heap_sort_time, marker='o', label='Heap Sort(Time)', color='c')
ax2.plot(n_values, tim_sort_time, marker='o', label='Tim Sort (Time))', color='k')
ax2.set_ylabel('Running Time(sec)', color='k')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

fig.tight_layout()
plt.title('Quick Sort vs. Merge Sort vs. Heap Sort')
plt.grid(True)

#  
plt.show()
```

**Optimized Merge Sort**

```python
import matplotlib.pyplot as plt

#  (n)
n_values = [100, 500, 1000, 5000, 10000, 50000, 100000, 200000, 500000]

# Merge Sort +,  
merge_sort_data = [1211, 8365, 18684, 117060, 254107, 1502659, 3205367, 6810267, 18312299]
merge_sort_time = [0.00029, 0.0019, 0.0042, 0.026, 0.057, 0.42, 0.93, 2.03, 5.9]

# 3_way merge Sort +,  
merge_sort_3_way_data = [1111, 7810, 17416, 109611, 236237, 1411075, 2999661, 6383659, 17242985]
merge_sort_3_way_time = [0.00017, 0.0012, 0.0026, 0.014, 0.028, 0.15, 0.3, 0.58, 1.56]

# Time Sort  
tim_sort_time = [9.81e-06, 4.61e-05, 9.52e-05, 0.0006, 0.0012, 0.007, 0.04, 0.03, 0.14]

#  
fig, ax1 = plt.subplots()

ax1.plot(n_values, merge_sort_data, marker='o', label='Merge Sort(C+S)', color='b')
ax1.plot(n_values, merge_sort_3_way_data, marker='o', label='3-ways M.Sort(C+S)', color='m')
ax1.set_xlabel('Input size (n)')
ax1.set_ylabel('comparison + swap', color='k')
ax1.set_yscale('log')  # +    

ax2 = ax1.twinx()
ax2.plot(n_values, merge_sort_time, marker='o', label='Merge Sort(Time)', color='r')
ax2.plot(n_values, merge_sort_3_way_time, marker='o', label='3-ways M.Sort(Time)', color='y')
ax2.plot(n_values, tim_sort_time, marker='o', label='Tim Sort (Time))', color='k')
ax2.set_ylabel('Running Time(sec)', color='k')

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

fig.tight_layout()
plt.title('Optimized Merge Sort')
plt.grid(True)

#  
plt.show()
```
