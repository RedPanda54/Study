# **알고리즘의 시간복잡도**

### ✅**좋은 알고리즘**의 기준


1.   정확성: **정확하게 기능**하는 알고리즘
2.   효율성: **수행시간은 짧고, 메모리 사용량이 적은** 효율적인 알고리즘
3.   로버스트성: 다양한 입력값에 대해 **범용적인 사용**이 가능한 알고리즘
4.   모듈화: 여러 가지 문제에 적용이 가능한 알고리즘
5.   가독성: 타인이 봐도 이해가 쉬운 알고리즘


## **Big-O 표기법**


*   어떤 알고리즘의 시간 복잡도를 나타내는 하나의 지표
*   최악의 수행시간을 나타냄으로서, 못해도 이정도의 성능을 지닌다는 것을 나타냄

### **특징**

1.   상수항을 무시
2.   계수를 무시
3.   ***최고차항만을 표기***

(ex) T(n) = 6n^2 + 4n + 3은 어떤 알고리즘의 수행시간 함수이다.

이것을 Big-O로 표기하면 다음과 같다.

> ▶ 최고차항을 제외한 (4n+3)을 무시하고 계수인 6을 무시 -> O(n^2)

## **수행시간비교**

같은 기능을 하는 두 함수로 수행시간의 차이를 직접 확인해보자.

```python
import time

# code for O(n^2)
def compute_e_ver1(n):
    for i in range(1, n+1):
        result = 1
        for j in range(1,i):
            result *= j
            X[i] = float(1/result)

# code for O(n)
def compute_e_ver2(n):
    Y[0] = 1
    result = 1
    for i in range(1,n):
        result *= i
        Y[i] = float(1/result)

n = int(input())

# index (n+1)과 n개의 리스트를 생성하고 함수를 호출해 덮어씌움
X = [0] * (n+1)
Y = [0] * n

# O(n^2)
before1 = time.process_time()
compute_e_ver1(n)
after1 = time.process_time()

Euler1 = 1
for i in range(1,n+1):
    Euler1 += X[i]

# O(n)
before2 = time.process_time()
compute_e_ver2(n)
after2 = time.process_time()

Euler2 = 0
for i in range(0,n):
    Euler2 += Y[i]

# 두 함수의 결과값을 확인
print('result of O(n^2) = ', Euler1)
print('result of O(n)   = ', Euler2)

# 두 함수의 수행시간 출력
print('running time of O(n^2) = ', after1-before1)
print('running time of O(n)   = ', after2-before2)
```

**대표 답안**

```python
import time, random

# code for O(n^2)-time version
def compute_e_ver1(n):
    result = 1
    for i in range(n):
        d = 1
        for j in range(0, i+1):
            d *= j + 1
        result += 1/d
        # time check
        after = time.process_time()
        if (after - before1) > 60:
            print('over time')
            break
    return result


# code for O(n)-time version
def compute_e_ver2(n):
    result = 1
    d = 1
    for i in range(n):
        d *= i + 1
        result += 1/d
        after = time.process_time()
        if (after - before2) > 60:
            print('over time')
            break
    return result

# n 입력받음
n = int(input("수 입력: "))

# compute_e_ver1 호출
before1 = time.process_time()
print(f'answer: {compute_e_ver1(n)}')
after1 = time.process_time()
print(f'time : {after1 - before1}')

# compute_e_ver2 호출
before2 = time.process_time()
print(f'answer: {compute_e_ver2(n)}')
after2 = time.process_time()
print(f'time : {after2 - before2}')
# 두 함수의 수행시간 출력
```

명확하게 시간 차이가 난다는 것을 볼 수 있고, n이 커짐에 따라 그 차이는 늘어날 것이다.

결론적으로 Big-O의 n의 함수 개형을 따져볼때, 차수가 높을 수록 수행시간이 오래 걸린다는 것을 알 수 있다.
